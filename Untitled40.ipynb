```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Setup for plots
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load Datasets
sentiment_df = pd.read_csv("/workspaces/trader-behavior-insights/trader-behavior-insights/data/fear_greed_index.csv")
trader_df = pd.read_csv("/workspaces/trader-behavior-insights/trader-behavior-insights/data/historical_data.csv")
```

```python
# Clean Column Names
sentiment_df.columns = sentiment_df.columns.str.strip().str.lower()
trader_df.columns = trader_df.columns.str.strip().str.lower()
```

```python
#  Rename Columns
sentiment_df.rename(columns={'date': 'date', 'classification': 'market_sentiment'}, inplace=True)
trader_df.rename(columns={'timestamp': 'time'}, inplace=True)
```

```python
# Format Date Columns
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'], errors='coerce')
trader_df['time'] = pd.to_datetime(trader_df['time'], errors='coerce')
trader_df['date'] = pd.to_datetime(trader_df['time'].dt.date, errors='coerce')
```

```python
# Drop null date rows
sentiment_df.dropna(subset=['date'], inplace=True)
trader_df.dropna(subset=['date'], inplace=True)
```

```python
#  Merge Datasets
merged_df = pd.merge(trader_df, sentiment_df, on='date', how='left')
```

```python
trader_df.rename(columns={'time': 'timestamp'}, inplace=True)
trader_df['Date'] = pd.to_datetime(trader_df['timestamp']).dt.date
trader_df['Date'] = pd.to_datetime(trader_df['Date'])
```

```python
# Check Merged Data
print("Merged DataFrame shape:", merged_df.shape)
print(merged_df[['date', 'market_sentiment']].drop_duplicates().head())
```

```python
#  Sentiment Distribution
sns.countplot(data=sentiment_df, x='market_sentiment', palette='viridis')
plt.title("Market Sentiment Distribution (Fear vs Greed)")
plt.show()
```
